```python
#https://python.hotexamples.com/es/examples/habanero/Crossref/-/python-crossref-class-examples.html
# https://habanero.readthedocs.io/en/latest/modules/crossref.html
def title2doi(title):
    title = title.lower()
    #clean_title = ''.join(e for e in title if e.isalnum())
    #print(clean_title)
    cr = Crossref()
    res = cr.works(query_title=title, select="title,DOI", limit=1)
    clean_title=" "
    doi=" "
    if(0<len(res['message']['items'])):
        fetched_title = res['message']['items'][0]['title']
        clean_title=" ".join(fetched_title)
        doi = res['message']['items'][0]['DOI']
    result=clean_title+"\t"+doi
    return result
    #return type(fetched_title)
    #return clean_title
    
 # Leer archivo de referencias
    # Import the os module
    #import os
    # Get the current working directory
    #cwd = os.getcwd()
    #cwd


#line = re.sub(r"</?\[\d+>", "", line)
g = open('data\ReferenciasWC.txt', "r")
#out = open('data\Results.csv', "r")

i=0
for line in g:
    if(i<800):
        n=line.split(")")[0].replace('(','') 
        x=line[::-1].split(";")[0][::-1].replace('et al.','')
        x=re.sub(r"[A-Z]\.", "", x).split(",")[1]
        title=x.split(".")[0]
          #print(title)
        doi=title2doi(title)
        total=n+"\t"+title+"\t"+doi+"\n"
        print(total)
        h = open("demofile.txt", "a", encoding="utf-8")
        h.write(total)
        h.close()
        i=i+1
g.close()   
```
